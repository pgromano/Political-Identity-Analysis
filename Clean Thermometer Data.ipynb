# Processing American National Election Survey 

The following script will load and process the ANES time-series cumulative data for public opinion from 1948-2015. Some of the variable names are non-intuitive so the variables are also renamed for clarity.

```python
survey = pd.read_table('../Data/anes_timeseries_cdf/anes_timeseries_cdf_rawdata.txt', sep='|')
```

```python
survey = survey.rename(columns = {'VCF0004':'Year', 'VCF0006':'ID', 'VCF0301':'Party'})
```

All respondents were asked about party affiliation with the options, "*Do you think of yourself as closer to the Republican or Democratic party?*", with the following provided options:

1.  Strong Democrat
2.  Weak Democrat
3.  Independent - Democrat        
4.  Independent - Independent
5.  Independent - Republican
6.  Weak Republican
7.  Strong Republican

A value of *0* was assigned for all respondents who had no party affiliation.

```python
survey = survey.replace({'Party' : { 1:'Strong Democrat', 
                           2:'Weak Democrat', 
                           3:'Independent Democrat',
                           4:'Independent',
                           5:'Independent Republican',
                           6:'Weak Republican',
                           7:'Strong Republican',
                           0:'No Party'}})
```

Additional survey questions were selected to identify respondent opinions on topics of religion, regions, unions, financial institution, and political terms such as Liberalism or Conservatism. Additionally, as the language deviates strongly overtime, we only analyzed data from 2012 and up to minimize impact.

```python
thermometer = survey[survey['Year'] >= 2012][
    ['Year','ID','Party','VCF0228','VCF0231',
     'VCF0290','VCF0211','VCF0218','VCF0212',
     'VCF0224','VCF0204','VCF0234','VCF0206',
     'VCF0207','VCF0217','VCF0227','VCF0219',
     'VCF0232','VCF0233','VCF0253','VCF0209',
     'VCF0210','VCF0210','VCF0220','VCF0223']]
```

```python
thermometer.to_csv('ANES_thermometer_survey.csv', index=False)
```
